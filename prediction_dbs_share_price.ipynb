```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/BC3415 AI in acc and finance/DBS_SingDollar.csv')
```

```python
X = df.loc[:,['SGD']]
Y = df.loc[:,['DBS']]
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
model = linear_model.LinearRegression()
model.fit(X,Y)
pred = model.predict(X)
print(mean_squared_error(Y, pred)**0.5)
```
